```python
#Embedding
import torch
from torch import nn as nn
import pandas as pd



def word_to_idx(x):
    L=[]
    for elem in x.split(' '):
    #eliminate \n and symbols, convert uppercase to lowercase
        elem=elem.replace('\n','').replace(',','').replace('.','').replace('”','').replace('?','').replace('\'','').lower().replace('“','').replace(']','').replace('[','')
        idx=Mylist.index(elem)
        L.append(idx)
    #print(L)
    print(len(L))
    return L


if __name__ == '__main__':


    filename='final_preproccessed_data.csv'
    df=pd.read_csv(filename)
    print(df.columns)
    print(df.shape)
    # df=df[0:100]

    # calculate total number of words
    text = df['text'].values
    mydict = {}
    for content in text:
        for elem in content.split(' '):
            elem = elem.replace('\n', '').replace(',', '').replace('.', '').replace('”', '').replace('?', '').replace(
                '\'', '').lower().replace('“', '').replace(']', '').replace('[', '')
            if elem not in mydict:
                mydict[elem] = 1
    print(len(mydict))

    Mylist = []
    for k, v in mydict.items():
        Mylist.append(k)
    print(Mylist[1:100])


    embedding = nn.Embedding(len(Mylist), 128)
    def list_to_vec(L):
        input = torch.LongTensor(L)
        RES = embedding(input).detach().numpy()
        print(RES.shape)
        return RES


    print('#######################1')
    df['embedding_idx'] = df['text'].map(lambda x: word_to_idx(x))

    print('#######################2')
    df['embedding_vec'] = df['embedding_idx'].map(lambda x: list_to_vec(x))

    df.to_csv('Pytorch_embeding.csv',header=True)
```
